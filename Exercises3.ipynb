### Step 9. Get the last business day of each month

```python
from pandas.tseries.offsets import BusinessMonthEnd

dates = pd.date_range(start='2000-03-01', end='2020-12-31', freq='M')
last_business_days = pd.Series(dates).apply(lambda x: x + BusinessMonthEnd(0))
print(last_business_days)
```

```python
list[last_business_days]
```

### Step 10.  What is the difference in days between the first day and the oldest

```python
apple['Date'] = pd.to_datetime(apple['Date'])
oldest_date = apple['Date'].min()
newest_date = apple['Date'].max()
difference = (newest_date - oldest_date).days
difference
```

### Step 11.  How many months in the data we have?

```python
apple['YearMonth'] = apple['Date'].dt.to_period('M')
num_months = apple['YearMonth'].nunique()
num_months
```

### Step 12. Plot the 'Adj Close' value. Set the size of the figure to 13.5 x 9 inches

```python
plt.figure(figsize=(13.5, 9))
plt.plot(apple['Date'], apple['Close'], color='blue', marker='o', linestyle='-')

plt.title('Adj Close')
plt.xlabel('date')
plt.ylabel('Adj Close')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
