## 4. Indexing

- Create new arrays where you select every second element from xarray and yarray. Plot them on top of ```xarray``` and ```yarray```.

````python
new_xarrays = xarray[::2]
new_xarrays
````

````python
new_yarrays = yarray[::2]
new_yarrays
````

````python
plt.plot(xarray, yarray, color="#1A0115", linewidth=2, label='main array')  
plt.plot(new_xarrays, new_yarrays, color="#11FF50", marker='o', linestyle='--', markersize=4, label='new xarrays')
plt.grid(True)
plt.legend()
plt.show()
````

- Select all values of ```yarray``` that are larger than 0. Plot those on top of the regular ```xarray``` and ```yarray```plot.

````python
new_yarrays_2 = yarray[yarray > 0]
new_yarrays_2
````

````python
plt.plot(xarray, yarray, color="#BFF608", marker='s', markersize=8, label='main array')
plt.plot(new_yarrays_2, color="#1808F6", marker='o', markersize=8, label='new array')
plt.grid(True)
plt.legend()
plt.show()
````

````python
mask = yarray > 0
x_positive = xarray[mask]
y_positive = yarray[mask]

plt.plot(xarray, yarray, color="#4F0472", markersize=10, label='main array')
plt.scatter(x_positive, y_positive, color="#F61C08", label='y > 0')
plt.grid(True)
plt.legend()
````

## 5. Combining arrays

- Create an array filled with ones with the same shape as ```normal_array```. Concatenate it to ```normal_array``` along the first dimensions and plot the result:

````python
ones_array = np.ones((20,100), dtype = np.int16)
ones_array
````

````python
concatenated = np.concatenate((normal_array, ones_array), axis=1)
concatenated
````

````python
plt.plot(concatenated, color="#BFF608", marker='s', markersize=2)
plt.grid(True)
plt.legend()
plt.show()
````

````python
for i in range(concatenated.shape[0]):
    plt.plot(concatenated[i])

plt.title("Plot of Rows: Noise + Ones")
plt.grid(True)
plt.show()
````
